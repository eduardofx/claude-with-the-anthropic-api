## Initialize Client

Create an Anthropic client instance and specify the model to use.

```python
from anthropic import Anthropic
client = Anthropic()   
model = "claude-sonnet-4-6"
```

## Helper Functions

Define utility functions to manage conversation messages and interact with the Claude API. Note that the `chat()` function now accepts a `system` parameter to customize Claude's behavior.

```python
def add_user_message(messages ,text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)    

def chat(messages, system=None):
    message = client.messages.create(
        model=model,
        messages=messages,
        max_tokens=100,
        system=system
    )
    return message.content[0].text
```

## Example: Math Tutor with System Prompt

This example demonstrates how to use a system prompt to create a specialized math tutor that guides students through problems step-by-step rather than giving direct answers.

```python
# Start with an empty message list
messages = []

# Define a system prompt that shapes Claude's behavior
system = """
You are a patient math tutor.
Do not directly give the answer, but guide the student step by step.
Respond in Portuguese.
"""

# Ask Claude to help solve a math problem
add_user_message(messages, "How do I solve 5x+3=2?")
answer = chat(messages, system=system)

# Display the response
answer
```
