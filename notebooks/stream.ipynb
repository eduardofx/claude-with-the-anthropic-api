# Anthropic Streaming Examples

## Setup

```python
from anthropic import Anthropic

# Initialize the Anthropic client
client = Anthropic()

# Set up model and messages for examples
model = "claude-sonnet-4-6"
messages = [
    {"role": "user", "content": "Hello! Write a short poem about coding."}
]
```

## Simplified Text Streaming

Stream text from the API and print each chunk as it arrives.

```python
with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
```
